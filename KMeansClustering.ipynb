```python
# K Means algorithm is unsupervised machine learning technique 
# used to cluster data points.
# Error decreases as # of clusters increases
```

```python
import os
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\prafulla.kamble\Downloads\0PRK\ML\CodeBasicsRepo\ML\13_kmeans\income.csv')
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='blue')
plt.xlabel('Age')
plt.ylabel('Income($)')
# plt.legend('Age', 'Income($)')

```

```python
sn.scatterplot(df1['Age'],df1['Income($)'],
               hue=df1['Income($)'])
```

```python
# df1 and df3 clusters are not grouped correctly as 
# scaling is not correct(here Y axis)

# Scale featues Age and Income

scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df.drop('cluster',axis='columns',inplace=True)
df
```

```python
df['cluster'] = y_predicted
df
```

```python
km.cluster_centers_
# Centriod
```

```python
km.cluster_centers_[:,1]
```

```python
km.cluster_centers_[:,0]
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='blue')

# Centroids
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],
           color='purple',marker='*', label='Centroid')

plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
# Elbow plot
```

```python
k_rng = range(1,10)
sse = []  # Sum of Square Error

for k in k_rng:
#     print(k)
    km = KMeans(n_clusters=3)
    km.fit(df[['Age','Income($)']])
#     print("===>",df['Age'], df['Income($)'])
    sse.append(km.inertia_)
    
sse
```

```python
plt.plot(k_rng,sse)
```

```python
k_rng = range(1,10)
sse = []  # Sum of Square Error

for k in k_rng:
#     print(k)
    km = KMeans(n_clusters=4)
    km.fit(df[['Age','Income($)']])
#     print("===>",df['Age'], df['Income($)'])
    sse.append(km.inertia_)
    
sse
```

```python
plt.plot(k_rng,sse)
```
